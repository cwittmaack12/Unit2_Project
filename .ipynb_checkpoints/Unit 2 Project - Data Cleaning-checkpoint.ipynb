#### Data Dictionary - House Data

**id**:
> Description:  This is a specific identifier for each house in the data set
>
> Data type:  int64

**bedrooms**:
> Description:  Bedroom count for the house in the row.
>
> Data type:  int64

**bathrooms**:
> Description:  Bathroom count for house in the row.
>
> Data type:  float64

**sqft_living**:
> Description:  Total indoor livable square feet for the house in the row
>
> Data type:  int64  

**sqft_lot**:
> Description:  Total square feet of the lot for the home in the row.
>
> Data type:  float64

**floors**:
> Description:  Number of stores of the home in the row.
>
> Data type:  float64

**waterfront**:
> Description:  'Yes' if the home is waterfront and 'No' if it is not.
>
> Data type:  object 

**view**:
> Description:  Rates the view from the property rated as 'Very Poor', 'Good', 'Great', 'Neutral', nan, or 'Poor'
>
> Data type:  object 

**condition**:
> Description:  Rates the condition of the home as: 'OK Condition', 'Great Condition', 'Good Condition',
               'Very Bad Condition', 'Bad Condition'
>
> Data type:  object 

**grade**:
> Description:  An evaluation of the construction materials and level of craftsmanship used to build the house.
>
> Data type:  float64

**sqft_above**:
> Description:  Total livable square footage above grade.
>
> Data type:  float64

**sqft_basement**:
> Description:  Total square footage below grade.
>
> Data type:  int64  

**yr_built**:
> Description: Year that the home was built.
>
> Data type:  float64

**yr_renovated**:
> Description:  Most recent year that the home was renovated.
>
> Data type:  float64

**zipcode**:
> Description:  Zip code of the property.
>
> Data type:  float64

**lat**:
> Description:  Latitude
>
> Data type:  float64

**long**:
> Description:  Longitude
>
> Data type:  float64

**sqft_living15**:
> Description:  Average size of interior housing living spaces for the closest 15 houses.  Compares size of home to the 
                neighbors.
>
> Data type:  int64  

**sqft_lot15**:
> Description:  Average size of the lot for the closest 15 houses.  Compares size of lot to the 
                neighbors.
>
> Data type:  int64  

#### Data Dictionary - House Sales Data

**id**:
> Description:  Identifier for the home in the row.
>
> Data type:  int64 

**date**:
> Description:  Date the property last sold.
>
> Data type:  object

**price**:
> Description:  Price the home last sold for.
>
> Data type:  object
